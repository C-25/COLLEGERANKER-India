## Import Libraries

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Task 2: Visualizing the Target Distribution

```python
sns.displot(data['revenue'])
```

```python
sns.displot(np.log1p(data['revenue']))
```

## Task 3: Relationship between Film Revenue and Budget

```python
sns.scatterplot(data['revenue'],data['budget'])
plt.title("Relationship between Film Revenue And Budget")

plt.show()
```

```python
relation=data[['revenue','budget']].corr()
sns.heatmap(relation,vmin=0,vmax=1,center=0)
plt.show()
```

## Task 4: Distribution of Languages in Film

```python
data['original_language'].unique()
sns.stripplot(data['original_language'],data['revenue'])
plt.title('Distribution Of Languages in Film')
plt.show()
```

```python
data['original_language'].unique()
sns.barplot(data['original_language'],data['revenue'])
plt.title('Distribution Of Languages in Film')
plt.show()
```

```python
data['original_language'].unique()
sns.boxplot(data['original_language'],data['revenue'])

plt.title('Distribution Of Languages in Film')
plt.show()
```
